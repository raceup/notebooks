```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
REG_LIMIT = 75
NEG_LIMIT = -21
```

```python
def neg_torque(brake):
    if brake < REG_LIMIT:
        return brake * NEG_LIMIT / 10.0

    return (REG_LIMIT - (brake - REG_LIMIT) * REG_LIMIT / (100.0 - REG_LIMIT)) * NEG_LIMIT / 10.0
```

```python
brakes = np.arange(0, 100, 0.1)
torques = np.array([neg_torque(brake) for brake in brakes])

plt.plot(brakes, torques)
plt.show()
```
