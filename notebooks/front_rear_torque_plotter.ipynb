```python
import matplotlib.pyplot as plt
import numpy as np

torque_distribution_front_rear = np.arange(0, 10)
rear_torque = 2 / (1 + torque_distribution_front_rear)
front_torque = torque_distribution_front_rear * rear_torque

plt.plot(torque_distribution_front_rear, front_torque, label='front')
plt.plot(torque_distribution_front_rear, rear_torque, label='rear')
plt.legend()
plt.savefig('out.png')
%ls
```
